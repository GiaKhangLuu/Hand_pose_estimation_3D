```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from scipy.spatial.transform import Rotation as R

%matplotlib widget
```

```python
# Set up the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the origin
origin = np.array([0, 0, 0])

# Define the unit vectors
unit_x = np.array([1, 0, 0])
unit_y = np.array([0, 1, 0])
unit_z = np.array([0, 0, 1])

# Plot the unit vectors
ax.quiver(*origin, *unit_x, color='r', length=1, normalize=True)
ax.quiver(*origin, *unit_y, color='g', length=1, normalize=True)
ax.quiver(*origin, *unit_z, color='b', length=1, normalize=True)

# Set the limits of the plot
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])

# Label the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set the title
ax.set_title('Unit Vectors (X, Y, Z)')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Use the interactive backend
%matplotlib notebook

# Set up the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the origin
origin = np.array([0, 0, 0])

# Define the unit vectors
unit_x = np.array([1, 0, 0])
unit_y = np.array([0, 1, 0])
unit_z = np.array([0, 0, 1])

# Define the additional vector
vector = np.array([-0.05233596, 0, 0.99862953])

# Plot the unit vectors
ax.quiver(*origin, *unit_x, color='r', length=1, normalize=True)
ax.quiver(*origin, *unit_y, color='g', length=1, normalize=True)
ax.quiver(*origin, *unit_z, color='b', length=1, normalize=True)

# Plot the additional vector
ax.quiver(*origin, *vector, color='m', length=1, normalize=False)

# Set the limits of the plot
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])

# Label the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.show()
```
