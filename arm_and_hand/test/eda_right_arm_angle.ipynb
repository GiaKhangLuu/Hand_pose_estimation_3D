```python
import pandas as pd
import glob
import time
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
right_arm_angles_df = pd.read_csv("/home/giakhang/dev/pose_sandbox/data/2024-10-19/2024-10-19-15:49/right_arm_angle_2024-10-19-15:49.csv")

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 6))
axes = axes.flatten()

for i in range(6):
    ax = axes[i]
    #total_rows = len(angles_df)
    #start_idx = total_rows // 4  
    #end_idx = start_idx + total_rows // 2  
    #angles_df = angles_df.iloc[start_idx:end_idx]
    raw_joint_i_angles = right_arm_angles_df.iloc[:, i+2].values
    smooth_joint_i_angles = right_arm_angles_df.iloc[:, i+2+6].values
    ax.plot(raw_joint_i_angles)
    ax.plot(smooth_joint_i_angles)
    title = "Angle joint {}\nStd: {}".format(i + 1, round(raw_joint_i_angles.std(), 2))
    ax.set_title(title)
    ax.legend(['raw', 'smooth'])

plt.tight_layout()
plt.show()
```
