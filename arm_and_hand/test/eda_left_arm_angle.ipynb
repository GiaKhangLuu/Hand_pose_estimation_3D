```python
import pandas as pd
import glob
import time
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
angles_df = pd.read_csv("/home/giakhang/dev/pose_sandbox/data/2024-10-02/2024-10-02-16:54/arm_angle_2024-10-02-16:54.csv")

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 6))
axes = axes.flatten()

for i in range(6):
    ax = axes[i]
    #total_rows = len(angles_df)
    #start_idx = total_rows // 4  
    #end_idx = start_idx + total_rows // 2  
    #angles_df = angles_df.iloc[start_idx:end_idx]
    raw_joint_i_angles = angles_df.iloc[:, i+2].values
    smooth_joint_i_angles = angles_df.iloc[:, i+2+6].values
    ax.plot(raw_joint_i_angles)
    ax.plot(smooth_joint_i_angles)
    title = "Angle joint {}\nStd: {}".format(i + 1, round(raw_joint_i_angles.std(), 2))
    ax.set_title(title)
    ax.legend(['raw', 'smooth'])

plt.tight_layout()
plt.show()
```

```python
raw_joint1_angles = pid_angles_df["raw_joint_1"]
pid_joint1_angles = pid_angles_df["smooth_joint_1"]

plt.figure(figsize=(15, 6))

sns.lineplot(data=raw_joint1_angles, drawstyle="steps-pre", label="setpoint")
sns.lineplot(data=pid_joint1_angles, label='position')
#sns.lineplot(data=j6_v_history, label='velocity')
#sns.lineplot(data=j6_a_history, label='acceleration')

ticks, labels = plt.xticks()
plt.xlabel("time (s)")
plt.ylabel("position")
plt.legend()
plt.title("pid for Joint 1")
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 6))
axes = axes.flatten()

for i in range(6):
    ax = axes[i]
    raw_joint_i_angles = pid_angles_df["raw_joint_{}".format(i + 1)][600:]
    pid_joint_i_angles = pid_angles_df["smooth_joint_{}".format(i + 1)][600:]
    ax.plot(raw_joint_i_angles)
    ax.plot(pid_joint_i_angles)
    title = "Angle joint {}".format(i + 1)
    ax.set_title(title)
    ax.legend(['setpoint', 'current'])

plt.tight_layout()
plt.show()
```
