```python
# Initialize the figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Set the plot limits
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_zlim(-1, 1)

# Labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Create initial vectors
x_vec = np.array([1, 0, 0])  # Starting unit vector x
y_vec = np.array([0, 1, 0])  # Starting unit vector y

# Create a line for x and y vectors
line_x, = ax.plot([0, x_vec[0]], [0, x_vec[1]], [0, x_vec[2]], color='r', label='Vector X')
line_y, = ax.plot([0, y_vec[0]], [0, y_vec[1]], [0, y_vec[2]], color='b', label='Vector Y')

# Plot settings
ax.legend()
ax.set_title("Rotating Vectors")

# Update function for the animation
def update(frame):
    # Rotate vector x clockwise
    theta = np.radians(frame)
    x_vec = np.array([np.cos(theta), -np.sin(theta), 0])  # Rotating x in XY-plane (clockwise)

    # Calculate perpendicular vector y
    y_vec = np.array([np.sin(theta), np.cos(theta), 0])  # Perpendicular to x in XY-plane

    # Update data for x vector
    line_x.set_data([0, x_vec[0]], [0, x_vec[1]])
    line_x.set_3d_properties([0, x_vec[2]])

    # Update data for y vector
    line_y.set_data([0, y_vec[0]], [0, y_vec[1]])
    line_y.set_3d_properties([0, y_vec[2]])

    return line_x, line_y

# Create the animation
ani = FuncAnimation(fig, update, frames=range(0, 360, 2), interval=50, blit=True)

# Show the plot
plt.show()
```
