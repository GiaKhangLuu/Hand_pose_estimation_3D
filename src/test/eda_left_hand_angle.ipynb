```python
import pandas as pd
import glob
import time
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
left_hand_angles_df = pd.read_csv("data/2024-10-29/2024-10-29-16:17/left_hand_angle_2024-10-29-16:17.csv")

fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(20, 12))
axes = axes.flatten()

titles = ["THUMB_joint_1", "THUMB_joint_2", "THUMB_joint_3",
          "INDEX_joint_1", "INDEX_joint_2", "INDEX_joint_3",
          "MIDDLE_joint_1", "MIDDLE_joint_2", "MIDDLE_joint_3",
          "RING_joint_1", "RING_joint_2", "RING_joint_3",
          "PINKY_joint_1", "PINKY_joint_2", "PINKY_joint_3",]

for i in range(15):
    ax = axes[i]
    raw_joint_i_angles = left_hand_angles_df.iloc[:, i+2].values
    smooth_joint_i_angles = left_hand_angles_df.iloc[:, i+2+15].values
    ax.plot(raw_joint_i_angles)
    ax.plot(smooth_joint_i_angles)
    ax.set_title(titles[i])
    ax.legend(["raw", "smooth"])
    
plt.tight_layout()
plt.show()
```
